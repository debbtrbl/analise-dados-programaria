# Preparação / Feature Engineering


## Tratamento dos dados

```python
dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'].value_counts()
```

```python
dados = dados[dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'] == 'Empregado (CLT)'] # apenas serão utilizados os dados de CLT
```

```python
dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'].value_counts()
```

```python
dados['COR/RACA/ETNIA'].value_counts()
```

```python
lista_retirar = ['Prefiro não informar', 'Indígena', 'Outra'] # retirando pois são poucos dados que não representam o grupo como todo
```

```python
dados = dados[~dados['COR/RACA/ETNIA'].isin(lista_retirar)]
```

```python
dados['NAO_BRANCA'] = dados['COR/RACA/ETNIA'].apply(lambda x: 1 if x != 'Branca' else 0) # criando uma nova coluna que categoriza brancos e não brancos
```

```python
dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].value_counts()
```

```python
dados['TEMPO_EXPERIENCIA'] = dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].str.extract(r'(\d+)') # extrai apenas os números (anos de experiência) de uma coluna com respostas em texto
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts()
```

```python
dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.replace('.', '')
```

```python
dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.extract(r'(\d+)')
```

```python
dados['NUMERO DE FUNCIONARIOS'].value_counts(dropna=False)
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts(dropna=False) # verificando se tem nulos
```

```python
dados['TEMPO_EXPERIENCIA'] = dados['TEMPO_EXPERIENCIA'].fillna(0)
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts(dropna=False)
```

```python
dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].value_counts()
```

```python
dados['INSATISFACAO'] = 0
```

```python
dados.loc[dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].notnull(), 'Qual o principal motivo da sua insatisfação com a empresa atual?'].apply(lambda x: 1 if 'Salário' else 0)
```

```python
dados.loc[dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].notnull(), 'INSATISFACAO'] = dados.loc[dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].notnull(), 'Qual o principal motivo da sua insatisfação com a empresa atual?'].apply(lambda x: 1 if 'Salário' else 0)
```

```python
dados['INSATISFACAO'].value_counts() # todos os que estão insatisfeitos com salario são 1
```

```python
dados['NIVEL DE ENSINO'].value_counts()
```

```python
dados['NIVEL DE ENSINO'] = dados['NIVEL DE ENSINO'].apply(lambda x: 1 if x == 'Não tenho graduação formal ' else
                               1 if x == 'Estudante de Graduação' else
                               2 if x == 'Graduação/Bacharelado' else
                               3 if x == 'Pós-graduação' else
                               4 if x == 'Mestrado' else
                               5 if x == 'Doutorado ou Phd' else -1) # tranformando categoricas em numericas
```

## Separação dos dados

```python
dados.columns
```

```python
# colunas que serão utilizazdas

dados = dados[['IDADE', 'GENERO', 'NAO_BRANCA', 'TEMPO_EXPERIENCIA', 'INSATISFACAO',  'SETOR', 'REGIAO ONDE MORA',
       'NUMERO DE FUNCIONARIOS', 'SALARIO', 'NIVEL DE ENSINO', 'NOVO_NIVEL']]
```

```python
dados.columns
```

```python
dados = pd.get_dummies(dados, columns=['SETOR', 'REGIAO ONDE MORA', 'GENERO', 'NOVO_NIVEL'], drop_first = True)
```

```python
x = dados.drop('SALARIO', axis=1) # atributos
y = dados['SALARIO'] # target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

# Treinamento

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

# Avaliação

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel('Valores Reais')
plt.ylabel('Valores Previstos')
plt.title('Comparação entre Valores Reais e Previstos')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.show()
```
